# Heart Dataset EDA

## Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the dataset

```python
df = pd.read_csv("E:\projects\health_app\data\heart.csv")
df.head()
```

## Analysis

```python
df.info()
```

```python
# Statistical description
df.describe()
```

```python
# No. of patients with and without diabetes
sns.countplot(x='DEATH_EVENT', data=df)
DN, DY = df['DEATH_EVENT'].value_counts()   # DN = Death No, DY = Death Yes
print('Number of patients without diabetes: {}'.format(DN))
print('Number of patients with diabetes: {}'.format(DY))
```

```python
# Histogram
df.hist(figsize=(10,10))
plt.show()
```

```python
# Density plot
df.plot(kind='density', subplots=True, layout=(4,4), sharex=False , figsize =(20,20))
plt.show()
```

```python
# Box plot
df.plot(kind='box', subplots=True, layout=(4,4), sharex=False,sharey=False ,figsize =(20,20))
plt.show()
```

```python
# Pairplot
sns.pairplot(data=df, hue='DEATH_EVENT')
```

```python
# Separating x & y
y = df['DEATH_EVENT']
x = df.drop(df[['DEATH_EVENT']], axis=1)
```

```python
# Correlation heatmap
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```
